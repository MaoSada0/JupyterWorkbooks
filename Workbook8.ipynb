```python
#–ó–∞–¥–∞–Ω–∏–µ 3
for i in range(4):
    data_iris = iris.data[:, i].reshape(150, 1)
    for j in range(4):
        if i != j:
            data_iris = np.hstack([data_iris, iris.data[:,j].reshape(150, 1)])
            cluster = AgglomerativeClustering(n_clusters=3, linkage='ward')
            cluster.fit(data_iris)
            plt.figure(figsize=(10, 7))
            plt.scatter(data_iris[:, 0], data_iris[:, 1], c=cluster.labels_, cmap='rainbow')
```
