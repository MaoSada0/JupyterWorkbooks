```python
#4.3.2 задание


import numpy as np
import matplotlib.pyplot as plt

y_arr = []
x_arr = []
for x in range(1, 11):
    y = (np.sqrt(1 + np.exp(np.sqrt(x)) + np.cos(x) ** 2) / np.abs(1 - np.sin(x) ** 3)) + np.log(2 * x)
    y_arr.append(y)
    x_arr.append(x)

plt.plot(x_arr, y_arr)
plt.scatter(x_arr[:len(x_arr) // 2], y_arr[:len(y_arr) // 2])
plt.show()
```
