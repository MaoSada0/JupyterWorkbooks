```python
#3.3.2 задание

new_url = 'https://raw.githubusercontent.com/akmand/datasets/master/iris.csv'
data = pd.read_csv(new_url, on_bad_lines='skip')
data.head(10)
```

```python
#3.3.2 задание

data['sepal_length_cm'] = (data['sepal_length_cm'] - data['sepal_length_cm'].min()) / (data['sepal_length_cm'].max() - data['sepal_length_cm'].min())
data['sepal_width_cm'] = (data['sepal_width_cm'] - data['sepal_width_cm'].mean()) / data['sepal_width_cm'].std()
data.head(10)

```
