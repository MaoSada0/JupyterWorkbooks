```python
#1.3.1 Задание

import matplotlib.pyplot as plt
import numpy as np

fig = plt.figure()

ax = fig.add_subplot(111, projection='3d')

p1 = np.array([1, 2, 3])
p2 = np.array([3, 2, 1])
p3 = np.array([4, 5, 6])
p4 = np.array([6, 5, 4])

ax.scatter(p1[0], p1[1], p1[2])
ax.scatter(p2[0], p2[1], p2[2])
ax.scatter(p3[0], p3[1], p3[2])
ax.scatter(p4[0], p4[1], p4[2])

arr = [p1, p2, p3, p4]
arrName = ['x', 'y', 'z', 'w']

for i in range(len(arr) - 1):
    for j in range(i + 1, len(arr)):
        print(f"Расстояние Евклида между {arrName[i]} и {arrName[j]}: ", np.linalg.norm(arr[i] - arr[j]))
        print(f"Квадрат Евклида между {arrName[i]} и {arrName[j]}: ", np.linalg.norm(arr[i] - arr[j]) ** 2)
        print(f"Расстояние Чебышева между {arrName[i]} и {arrName[j]}: ", np.linalg.norm(arr[i] - arr[j], ord=np.inf))
        print(f"Расстояние Хемминга между {arrName[i]} и {arrName[j]}: ", np.linalg.norm(arr[i] - arr[j], ord=1))
        print()

plt.show()
```
